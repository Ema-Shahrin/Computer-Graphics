```python
import matplotlib.pyplot as plt
```

```python
# with respect to X-axis
x,y = 4,5
x1,y1 = 4,-5
```

```python
plt.scatter(x,y,color='b',marker='o',label='(x,y)')
plt.scatter(x1,y1,color='r',marker='o',label='(first x,first y)')

plt.axhline(0,color='black',linewidth=0.8)
plt.axvline(0,color='black',linewidth=0.8)

plt.legend()
plt.show()
```

```python
#x,y Axis
x,y = 4,5
x1,y1 = -4,-5

plt.scatter(x,y,color='b',marker='o',label='(x,y)')
plt.scatter(x1,y1,color='r',marker='o',label='(x1,y1)')

plt.axhline(0,color='k',linewidth=1.0)
plt.axvline(0,color='k',linewidth=1.0)

plt.legend()
plt.show()
```
